```
AccessType = alt.Chart(df).mark_bar().encode(
    alt.X("Access Type:O", bin=False),
    y='count():Q',
)

ContractLength = alt.Chart(df).mark_bar().encode(
    alt.X("Contract Length:O", bin=False),
    y='count():Q',
)

Technologies = alt.Chart(df).mark_bar().encode(
    alt.X("Technologies:O", bin=False),
    y='count():Q',
)

Unlimited = alt.Chart(df).mark_bar().encode(
    alt.X("Unlimited Data:O", bin=False),
    y='count():Q',
)

Support247 = alt.Chart(df).mark_bar().encode(
    alt.X("24x7 Support:O", bin=False),
    y='count():Q',
)

StaticIP = alt.Chart(df).mark_bar().encode(
    alt.X("Static IP:O", bin=False),
    y='count():Q',
)

RouterInc = alt.Chart(df).mark_bar().encode(
    alt.X("Router Included:O", bin=False),
    y='count():Q',
)

AccessType | ContractLength | Technologies | Unlimited | Support247 | StaticIP | RouterInc
```

```

lst = ['Unlimited Data','24x7 Support','Static IP','Router Included']  

percentlist = []

for n in lst:
    CountTrue = 0
    CountFalse = 0
    for i in df.index:
        if df.at[i, n] == True:
            CountTrue += 1
        else:
            CountFalse += 1

    percentTrue = (CountTrue / (CountTrue + CountFalse)) * 100
    percentlist.append(percentTrue)

print(percentlist)
```

```
alt.Chart(df).mark_bar().encode(
    alt.X("Mbps Down:O", bin=False),
    y='count():Q',
)
```

```
categories = ['0-25','26-50','51-100','101-150','151-250','251-500','>500']

df['Mbps Down (categories)'] = ''

for i in df.index:
    if df.at[i, 'Mbps Down'] < 26:
        df.at[i, 'Mbps Down (categories)'] = categories[0]
    
    elif df.at[i, 'Mbps Down'] > 25 and df.at[i, 'Mbps Down'] < 51:
        df.at[i, 'Mbps Down (categories)'] = categories[1]
    
    elif df.at[i, 'Mbps Down'] > 50 and df.at[i, 'Mbps Down'] < 101:
        df.at[i, 'Mbps Down (categories)'] = categories[2]
    
    elif df.at[i, 'Mbps Down'] > 100 and df.at[i, 'Mbps Down'] < 151:
        df.at[i, 'Mbps Down (categories)'] = categories[3]
    
    elif df.at[i, 'Mbps Down'] > 150 and df.at[i, 'Mbps Down'] < 251:
        df.at[i, 'Mbps Down (categories)'] = categories[4]
    
    elif df.at[i, 'Mbps Down'] > 250 and df.at[i, 'Mbps Down'] < 501:
        df.at[i, 'Mbps Down (categories)'] = categories[5]
    
    elif df.at[i, 'Mbps Down'] > 500:
        df.at[i, 'Mbps Down (categories)'] = categories[6]

df[['Mbps Down','Mbps Down (categories)']].sample(10)
```

```
DownSpeed = alt.Chart(df).mark_bar().encode(
    x = alt.X("Mbps Down (categories):O", bin=False,
              sort = categories),
    y='count():Q',
)

MRCLocal = alt.Chart(df).mark_bar().encode(
    x = alt.X("Mbps Down (categories):O", bin=False,
    sort = categories),
    y='average(MRC Local)',
)

DownSpeed | MRCLocal
```
