```
alt.Chart(df).mark_bar().encode(
    alt.X("Mbps Down:O", bin=False),
    y='count():Q',
)
```

```
categories = ['0-25','26-50','51-100','101-150','151-250','251-500','>500']

df['Mbps Down (categories)'] = ''

for i in df.index:
    if df.at[i, 'Mbps Down'] < 26:
        df.at[i, 'Mbps Down (categories)'] = categories[0]
    elif df.at[i, 'Mbps Down'] > 25 & df.at[i, 'Mbps Down'] < 51:
        df.at[i, 'Mbps Down (categories)'] = categories[1]
    elif df.at[i, 'Mbps Down'] > 50 & df.at[i, 'Mbps Down'] < 101:
        df.at[i, 'Mbps Down (categories)'] = categories[2]
    elif df.at[i, 'Mbps Down'] > 100 & df.at[i, 'Mbps Down'] < 151:
        df.at[i, 'Mbps Down (categories)'] = categories[3]
    elif df.at[i, 'Mbps Down'] > 150 & df.at[i, 'Mbps Down'] < 251:
        df.at[i, 'Mbps Down (categories)'] = categories[4]
    elif df.at[i, 'Mbps Down'] > 250 & df.at[i, 'Mbps Down'] < 501:
        df.at[i, 'Mbps Down (categories)'] = categories[5]
    elif df.at[i, 'Mbps Down'] > 500:
        df.at[i, 'Mbps Down (categories)'] = categories[6]

df[['Mbps Down','Mbps Down (categories)']].sample(10)
```
