```
#import necessary libraries
import pandas as pd
import numpy as np
import altair as alt
```

```
current_link = "Europe_AM.csv"
current_df = pd.read_csv(current_link)
current_df.head(20)
```

```
past_link = 'europe_2018-19.csv'
past_df = pd.read_csv(past_link)
past_df.head(20)
```

```
#rename key columns in past dataframe to match present dataframe
past_df = past_df.rename(columns = {'Date':'Quarter','Mbps (Download)':'Mbps Down', 'Tariff Name':'Tariff', 'Access':'Access Type',
                                    'Unlimited Data Allowance':'Unlimited Data', 'One-Time Fee (Local)':'NRC Local', 'MRC (Local)':'MRC Local',
                                    'One-Time Fee (USD)':'NRC USD', 'MRC (USD)':'MRC USD','Local Currency':'Currency'})

past_df.head(10)
```

```
#remove commas and dollar signs from price and speed columns

int_list = ['Mbps Down','MRC Local', 'NRC Local', 'MRC USD', 'NRC USD']

for i in int_list:
    past_df[i] = past_df[i].str.replace(',', '').str.replace('$', '').astype(float)

past_df.head(10)
```

```
#combine old and new dataframes (the number of rows will not match the index, needs to be adjusted)
combine_df = pd.concat([current_df, past_df], axis=0)
combine_df
```

```
#adjust index after merging dataframes(tables)
combine_df = combine_df.reset_index()
combine_df = combine_df.drop(combine_df.columns[[0]], 1)
combine_df
```

```
categories = ['0-25','26-50','51-100','101-150','151-250','251-500','>500']

document = combine_df

document['Mbps Down (categories)'] = ''

for i in document.index:
    if document.at[i, 'Mbps Down'] < 26:
        document.at[i, 'Mbps Down (categories)'] = categories[0]
    
    elif document.at[i, 'Mbps Down'] > 25 and document.at[i, 'Mbps Down'] < 51:
        document.at[i, 'Mbps Down (categories)'] = categories[1]
    
    elif document.at[i, 'Mbps Down'] > 50 and document.at[i, 'Mbps Down'] < 101:
        document.at[i, 'Mbps Down (categories)'] = categories[2]
    
    elif document.at[i, 'Mbps Down'] > 100 and document.at[i, 'Mbps Down'] < 151:
        document.at[i, 'Mbps Down (categories)'] = categories[3]
    
    elif document.at[i, 'Mbps Down'] > 150 and document.at[i, 'Mbps Down'] < 251:
        document.at[i, 'Mbps Down (categories)'] = categories[4]
    
    elif document.at[i, 'Mbps Down'] > 250 and document.at[i, 'Mbps Down'] < 501:
        document.at[i, 'Mbps Down (categories)'] = categories[5]
    
    elif document.at[i, 'Mbps Down'] > 500:
        document.at[i, 'Mbps Down (categories)'] = categories[6]

document[['Mbps Down','Mbps Down (categories)']].sample(10)
```
